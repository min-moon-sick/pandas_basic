```python
# 데이터프레임의 행과 열을 바꿈 : 전치행렬
import pandas as pd
```

```python
exam_data = {'이름' : ['서준','우현','인하'],
            '수학' : [90, 80, 70],
            '영어' : [98,89,95],
            '국어' : [98,89,95]}

df = pd.DataFrame(exam_data)

# df.set_index(['국어', '영어'], inplace=True)
# df.set_index(['이름'], inplace=True)
# df
```

```python
df.T
```

```python
# reindex(새로운 인덱스 배열) : 기존의 자료변경 안함, 새로운 데이터프레임 반환
dict_data = {'c0' : [1,7,2],
            'c1' : [1,3,4],
            'c2' : [2,8,5],
            'c3' : [4,5,2]}

df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])

new_index = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf = df.reindex(new_index)

print(ndf ,'\n')

# 존재하지 않는 인덱스의 값을 0으로 채워줌
ndf1 = df.reindex(new_index, fill_value = 0)
print(ndf1)
```

```python
# reset_index() : 인덱스 초기화
print(df, '\n')
print(df.reset_index())     # 기존 인덱스는 컬럼으로 변환, 인덱스는 초기화

df.reset_index(drop=True)  # 기존 인덱스를 삭제, 인덱스는 초기화

print('\n', df)
```

```python
# 인덱스를 기준으로 정렬
df.sort_index(ascending=False)
```

```python
# 칼럼을 기준으로 정렬
df.sort_values(by = ['c0', 'c2'], ascending=False)
```

```python
# 시리즈 연산 : 시리즈 +,*,/,- 사칙연산 가능
# 1. 인덱스로 정렬
# 2. 매칭되는 값에 대해 연산, Nan이 있으면 결과는 Nan
# 3. 결과 리턴 -> 인덱스로 정렬된 상태
```

```python
student = pd.Series({'국어' : 100,
                    '수학' : 90,
                    '영어' : 80,})

print(student, '\n')
print(student + 50 , '\n')
```

```python
# 시리즈 연산
import numpy as np

student1 = pd.Series({'국어' : np.nan,
                    '수학' : 90,
                    '영어' : 80,})

student2 = pd.Series({'수학' : np.nan,
                    '국어' : 90,
                    '영어' : 80,
                    '과학' : 80})

print(student1 + student2)
```

```python
# 시리즈 연산 메서드 : 시리즈.add(시리즈1, fill_value=0)   nan을 0으로 채워줌
print(student1.add(student2, fill_value=0), '\n')
print(student1.sub(student2, fill_value=0), '\n')
print(student1.mul(student2, fill_value=0), '\n')
print(student1.div(student2, fill_value=0), '\n')
```
