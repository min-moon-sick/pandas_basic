```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic = sns.load_dataset('titanic')

# age와 fare 가져오기
df = titanic.loc[:, ['age', 'fare']]
df.info()
```

```python
# 숫자에 10을 더한 값을 구하는 함수
def add_10(n):
    return n + 10

# 두 객체의 합을 구하는 함수
def add_obj(a, b):
    return a + b
```

```python
# age 칼럼의 값에 10을 더한 값을 실행
df['age+10'] = df['age'].apply(add_10)

# age와 fare의 값을 더함
df['age+20'] = df['age'].apply(add_obj, b= 20)
df.head()

df['age'].apply(lambda x: x + 1)
df
```

```python
# df 모든 원소에 함수 적용
df.applymap(add_10)
```

### 데이터프레임 객체에 함수 매핑

```python
# 데이터프레임.pipe(함수) 

def missing_value(x):
    return x.isnull()

def missing_cnt(x):
    return missing_value(x).sum()

def total_num_missing(x):
    return missing_cnt(x).sum()
```

```python
result_value = df.pipe(missing_value)
print(type(result_value))
print(result_value)      # 데이터프레임

result_value = df.pipe(missing_cnt)
print(type(result_value))
print(result_value)      # 개수

result_value = df.pipe(total_num_missing)
print(type(result_value))
print(result_value)      # 값
```

```python
# 필터링 :불린 인덱싱을 사용
# 데이터프레임[불린 인덱싱] : True인 행만 가져옴

titanic = sns.load_dataset('titanic')

# 나이가 10대인 자료만 추출
titanic.loc[ (10<= titanic['age']) & (titanic['age'] < 20), :]

# 나이가 10대 미만이고 여성만 자료 추출
titanic.loc[ (titanic['sex'] == 'female') & (titanic['age'] < 10), :]

# 나이가 60대이상이고 생존자중, age, sex, class, alive 컬럼만 추출
titanic.loc[ (titanic['survived'] == True) & (60<= titanic['age']), :][['age', 'sex', 'class', 'alive']]
```

```python
# sibsp 컬럼 : 형제 또는 배우자의 수 -> 수가 3,4,5인 승객의
# age, class, sibsp, alive 컬럼만 추출

titanic.loc[(titanic.sibsp >= 3) & (titanic.sibsp < 6), :][['age', 'class', 'sibsp', 'alive']].head(3)
```

```python
# titanic 데이터를 로드해 Nan 데이터를 처리 : age는 평균값으로, deck 칼럼 삭제
# 다른 컬럼은 이전 값으로 대체
# 평균 age보다 적은 age를 가진 승객 중에 여성과 생존자 출력
# age, sex, alive, class 칼럼만 추출
# 가족의 수가 4~5명인 승객의 나이, 가족수, 생존여부 추출

df = sns.load_dataset('titanic')

# nan 개수 확인
print(df.isnull().sum())

# age nan에 평균값 채우기
df.age.fillna(df.age.mean(), inplace= True)

# deck 컬럼 삭제
df = df.drop('deck', axis = 1)

# 결측값 이전값으로 채우기
df.ffill(inplace=True)

# 평균 age보다 적은 age를 가진 승객 중에 여성과 생존자 출력
df.age.mean()  # 29.69911764705882
df.loc[ (df.age < df.age.mean())].head(3)

# age, sex, alive, class 칼럼만 추출
df.loc[ (df.age < df.age.mean())][['age', 'sex', 'alive', 'class']].head(3)

# 가족의 수가 4~5명인 승객의 나이, 가족수, 생존여부 추출
df.loc[ (df.sibsp <= 5) & (df.sibsp >= 4)][['age', 'sibsp', 'alive']].head(3)
```

#### 데이터 프레임 합치기

```python
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
```

```python
# pd,concat(데이터프레임 리스트)
result1 = pd.concat([df1, df2])

result2 = pd.concat([df1, df2], ignore_index=True)

result3 = pd.concat([df1, df2], axis=1)

# nan이 없고 공통에 인덱스를 갖고 있는 부분만 concat
result4 = pd.concat([df1, df2], axis=1, join='inner')
```
